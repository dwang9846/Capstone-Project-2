```python
#check to see how many missing value each feature has
data.isnull().sum()
```

```python
#check to see which feature has more than 20 missing value
data.isnull().sum(axis=1)>20
```

```python
#extract meaningful data out of original dataset
#keep features whose missing value are less than 10,000 in a column
#keep records whose missing value are less than 40 in a row 
data1=data.loc[data.isnull().sum(axis=1)<40,data.isnull().sum()<10000]
```

```python
#get the new dataset data1 and remove the 1st feature('id') since no meaningful information 
data1.ix[:,2:]
```
